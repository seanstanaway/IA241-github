## Introduction and Background of Covid-19

The Covid-19 disease, also known as the coronavirus disease 2019, is a virus and was first discovered in December 2019. This disease is known to be extremely contagious and quickly spread throughout the world. The disease most commonly causes respiratory symptoms that are similar to the feeling of a cold, flu, or pneumonia. This led to a worldwide pandemic and lockdown, hospitalizations, and many deaths. Countries economies took a massive hit and are still recovering. Countries around the world are still experiencing the effects of Covid-19 even after over two years later. The virus is spread easily from person to person by close contact with someone infected. It can also be spread through contact with droplets from an infected person's cough, or sneeze. Covid-19 was caused by a virus called SARS-CoV-2, which belongs to the coronavirus family, that can result in many different diseases from head or chest colds to more severe ones. 

Information from CDC (Center for Disease Control and Prevention)

## Canada's Covid-19 Data

This is a small-scale amount of Canada's Covid-19 data, that was collected from the ECDC. The data displays the top ten rows of the spreadsheet. It was pulled by using "canada_data" to identify it from "countriesandTerritories" and using the df.loc function. 

```python
canada_data = df.loc[ df['countriesAndTerritories']=='Canada' ]

canada_data[:10]
```

## Month with the Highest Number of Deaths from Covid-19 in 2020

```python
sum_death_by_month = canada_data.groupby('month').sum()['deaths']
sum_death_by_month.plot.bar()
```

### Explanation of Calculation
The code shows the number of deaths by month in the year 2020, in Canada, due to Covid-19. This is done by grouping the sum of 'month' and 'deaths' and using the variable sum_death_by_month, in order to show the sum of deaths by month. I chose to use a bar chart because I felt it would be the most effective to display this data. I did this by adding.plot.bar() to the end of my code, after "sum_death_by_Month"

### Interpretation of Visualization
As seen in the bar graph, the month with the highest number of deaths from Covid-19 in 2020 in Canada was May. The number of deaths reached up to approximately 4,000. The month with the second highest number of deaths is April, which comes right before May. This month got up to about 3,000 deaths. In one month, the number of deaths jumped by 1,000. In March, right before April, there were way less than 500 deaths, meaning in between those two months there was an increase of a little less than 4,000. From May to June there was a large decrease of about 2,500 deaths and from June until September there was a steady decrease. The number of deaths in Canada did not rise again until October. 

## How the Number of Cases is Related to the Number of Deaths

```python
canada_data.plot.scatter(x='cases',y='deaths')
```

### Explanation of Calculation

This code shows the relation of the number of cases to the number of deaths in Canada. In order to represent this in the most effective way I used a scatter plot, by utilizing "canada_data.plot.scatter". I then labeled the axis'. The x axis is labeled as cases (x ='cases') and y axis is labeled as deaths (y='deaths').

### Interpretation of Visualization

It can be seen in the scatter plot that as time increased, in months, both cases and deaths did. As moving from left to right in the visual, there is an uphill pattern. This shows a positive relationship between the x variable, cases, and y variable, deaths. As the number cases moves right, increases, the number of deaths tend to move up, also increasing. This is not a linear pattern. The highest number of cases can be noticed at about 8,000, where deaths go above 200, from this dataset. It can be interpreted this pattern is continued as cases and deaths increase. There are also a few outliers that can be recognized in the data. 


## The Trend in Number of Confirmed Cases from January 2020 to December 2020

```python
canada_case_by_date = df.groupby('dateRep').sum()['cases']
canada_case_by_date.plot()
```

### Explanation of Calculation

This code identifies the trend in number of confirmed covid cases from January 2020 to December 2020. I calculated this by grouping the sum of dateRep, ('dateRep'), and of the cases, ['cases'], and then appointed them to "canada_case_by_date". For this question I chose to use a line graph because I knew that would present the data in the most constructive way. I did this by using the variable "canada_case_by_date" while incorporating ".plot(), so it would exhibit the graph. 

### Interpretation of Visualization

It can be noticed in this visualization there are multiple peaks at majority of the months. There are multiple spikes in cases. The highest peak reaches just about 700,000 cases. The relationship in the linegraph is known to be non-monotonic because the lines do not only go up or down, it is more complex. It is recognized that the first wave of Covid-19 began to infect Canada in January 2020. For the most part the cases stayed consistently bad throughout the year. The number of cases never dropped lower than 500,000 in this year. This shows how hard Canada got hit with the infectious disease throughout the year 2020.
